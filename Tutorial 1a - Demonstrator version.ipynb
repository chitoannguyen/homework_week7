```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### shape of the arrays
d=(7, 7,3)
### define zero arrays
arrayD=np.zeros(d)
arrayE=np.zeros(d)
arrayF=np.zeros(d)
### color index
red = [1,0,0];
white = [1,1,1];
black = [0,0,0];
green = [0,1,0];
blue  = [0,0,1];
```

```python
#arrayD
for i in range(0,np.shape(arrayD)[0]):
    for j in range(0,np.shape(arrayD)[1]):
            if ((i+j)%2)==0:
                arrayD[i,j] = red
            else:
                arrayD[i,j] = blue
plt.imshow(arrayD, interpolation='none')
plt.title('arrayD')
```

```python
#arrayE
for i in range(0,np.shape(arrayE)[0]):
    for j in range(0,np.shape(arrayE)[1]):
            if (i%2)==0:
                arrayE[i,j] = red
            else:
                arrayE[i,j] = blue

plt.imshow(arrayE, interpolation='none')
plt.title('arrayE')
```

```python
#arrayF
for i in range(0,np.shape(arrayF)[0]):
    for j in range(0,np.shape(arrayF)[1]):
            if (2<=i<=4) and (2<=j<=4):
                arrayF[i,j] = red
            else:
                arrayF[i,j] = blue

plt.imshow(arrayF, interpolation='none')
```
